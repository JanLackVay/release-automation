```python
import time 

print("Check queued jobs:")
queue_response = requests.get(
    "http://jenkins-ber.reeinfra.net/queue/api/json",
    auth=(credentials["username"], credentials["credentials"]),
).json()

queued_items = []
if queue_response["items"]:
    for item in queue_response["items"]:
        if (
            item["actions"][0]["parameters"][0]["value"]
            == "release-testbed-validation-gamma"
        ):
            print(f"Canceling job {item['id']}")
            queued_items.append(item["id"])
            response = requests.post(
                f"http://jenkins-ber.reeinfra.net/queue/cancelItem?id={item['id']}",
                auth=(credentials["username"], credentials["credentials"]),
            )
            print(response.status_code)
            print(response.text)
start_time = time.time()
timeout = 60
while queued_items:
    if time.time() - start_time > timeout:
        print("Timeout: Unable to clear the queue within 60 seconds.")
        break
    time.sleep(1)
    queue_response = requests.get(
        "http://jenkins-ber.reeinfra.net/queue/api/json",
        auth=(credentials["username"], credentials["credentials"]),
    ).json()
    queued_item_ids = [item["id"] for item in queue_response["items"]]

    # Check if any queued items from your list are still present
    queued_items = [
        item_id for item_id in queued_items if item_id in queued_item_ids
    ]
    print(f"Queued items: {queued_items}")
    if not queued_items:
        print("All canceled jobs have been removed from the queue.")
        break

print("\nDeployment started")
```
